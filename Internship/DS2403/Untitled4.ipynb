```python
from bs4 import BeautifulSoup
import requests
```

```python
Page = requests.get ('https://www.cnbc.com/world/?region=world%20an')
Page
```

```python
Soup= BeautifulSoup (Page.content)
Soup
```

```python
First_title = Soup.find('div',class_="Card-titleContainer")
First_title
```

```python
from datetime import datetime
```

```python
sta = Soup.find('span',class_="Card-time")
sta
```

```python
link= Soup.find("href",class_="Card-time")
link
```

```python
News_link = []
for i in Soup.find_all("href",class_="Card-titleContainer"):
    News_link.append(i['data-src'])
News_link
```

```python
Headings =[]
for i in Soup.find_all('div',class_="Card-titleContainer"):
    Headings.append(i.text)
Headings
```

```python
Date =[]
for i in Soup.find_all('span',class_="Card-time"):
    Date.append(i.text)
Date
```

```python
print(len(Headings),len(Date))
```
